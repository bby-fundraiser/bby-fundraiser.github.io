```python
csv_path = "/Users/salvatoregiorgi/Downloads/auction_items.csv"
csv_path = "/Users/salvatoregiorgi/Downloads/BBY 25 QUIZZO SPRING FUNDRAISER - posts.csv"
with open(csv_path, "r") as f:
    reader = csv.reader(f, delimiter=",")
    header = next(reader)
    name_idx = header.index("name")
    desc_idx = header.index("description")
    website_idx = header.index("website")
    image_idx = header.index("image")
    keep_idx = header.index("keep")
    sb_idx = header.index("starting bid")
    other_images_idx = header.index("other images")

    current_date = datetime.utcnow()

    for i, line in enumerate(reader):
        if line[keep_idx] == "n": continue
        name = line[name_idx].strip()
        if not name: continue
        desc = line[desc_idx] + "."
        
        if line[website_idx]:
            image_link = """[Website](""" + line[website_idx] + """)"""
        else:
            image_link = ""
        if line[image_idx]:
            image_name = line[image_idx]
        else:
            image_name = name.lower().replace(" ", "_")
        
        formatted_date = current_date.strftime("%Y-%m-%d 12:00:00 +0000")

        
        this_text = post_text.format(
            name=name, 
            date=formatted_date,
            desc=desc, 
            image_name=image_name,
            image_link=image_link,
            )
        
        if line[sb_idx]:
            this_text += """\nStarting bid: """ + line[sb_idx]

        if line[other_images_idx]:
            names = [i.strip() for i in line[other_images_idx].split(";")]
            this_text += "\n"
            for n in names:
                this_text += """\n![{n}](/assets/img/{n})""".format(n=n)
        
        if name == "Reanimator":
            image_link = """<ul>
            <li>[Ceramic Mug](https://www.reanimatorcoffee.com/products/ceramic-mug)</li>
            <li>[Tote bag](https://www.reanimatorcoffee.com/products/tote-bag)</li>
            <li>[Can glass](https://www.reanimatorcoffee.com/products/reanimator-can-glass)</li>
            <li>and more!</a>
        </ul>"""


        file_name = """/Users/salvatoregiorgi/Documents/workspace/bby-fundraiser.github.io/_posts/{d}-{fname}.markdown""".format(
            d = current_date.strftime("%Y-%m-%d"),
            fname=name.lower().replace(" ", "-").replace("&", "").replace("'","")
            )
        with open(file_name, "w") as file:
            file.write(this_text)

        current_date -= timedelta(days=1)
```
